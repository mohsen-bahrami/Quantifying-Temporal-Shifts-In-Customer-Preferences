```python
import pandas as pd
import os
import shutil
```

```python
year = 2021

norm_ranges = list(range(2, 31))
PSO_ranges = list(range(5, 75, 5))
```

```python
directory = 'norm_PSO_combination_' + str(year)
fnished_folders = os.listdir(directory)
folders = [i for i in fnished_folders if '_PSO_' in i]
```

```python
len(folders)
```

```python
def process_one_combination(combination):
    norm_here = combination.split('_')[1]
    PSO_range_here = combination.split('_')[-1]
    path_here = directory + '/' + combination + '/PSO_' + str(year) + '_6params_NYC_' + combination + '.csv'
    dest_dire = 'data/3_PSO_results/PSO_calirabtion_results/'
    copied_dire =  dest_dire + 'norm_PSO_combination_'+str(year) + '/' + 'PSO_' + str(
        year) + '_6params_NYC_' + combination + '.csv'
    shutil.copy(path_here, copied_dire)
    
    df_temp = pd.read_csv(path_here)
    if df_temp.shape[0] != 6493 or df_temp['cbg'].nunique() != 6493:
        print(combination, '?')
    df_temp = df_temp[df_temp['cost'] != 'no visitors']
    df_temp['cost'] = df_temp['cost'].astype(float)
    return int(norm_here), int(PSO_range_here), df_temp['cost'].mean(), df_temp['cost'].median()
```

```python
error_stats = {'norm': [], 'PSO_range': [], 'mean': [], 'median': []}

for indx, f in enumerate(folders):
    values_here = process_one_combination(f)
    error_stats['norm'].append(values_here[0])
    error_stats['PSO_range'].append(values_here[1])
    error_stats['mean'].append(values_here[2])
    error_stats['median'].append(values_here[3])
    if (indx+1) % 40 == 0:
        print("{:.2%}".format((indx+1)/len(folders)), ' finished')
```

```python
error_stats_df = pd.DataFrame(error_stats).sort_values(by = ['norm', 'PSO_range'])
error_stats_df
```

```python
error_stats_matrix_mean = {}
error_stats_matrix_median = {}

for n in norm_ranges:
    error_stats_matrix_mean[n] = {}
    error_stats_matrix_median[n] = {}

for indx, cont in error_stats_df.iterrows():
    error_stats_matrix_mean[cont['norm']][int(cont['PSO_range'])] = cont['mean']
    error_stats_matrix_median[cont['norm']][int(cont['PSO_range'])] = cont['median']
```

```python
error_stats_matrix_mean_df = pd.DataFrame(error_stats_matrix_mean).T
error_stats_matrix_mean_df
```

```python
error_stats_matrix_median_df = pd.DataFrame(error_stats_matrix_median).T
error_stats_matrix_median_df
```

```python
error_stats_matrix_mean_df.to_csv(
    'data/3_PSO_results/norm_PSO_simulated_annealing_data/norm_PSO_error_means_'+str(
        year)+'.csv')
```

```python
error_stats_matrix_median_df.to_csv(
    'data/3_PSO_results/norm_PSO_simulated_annealing_data/norm_PSO_error_medians_'+str(
        year)+ '.csv')
```
