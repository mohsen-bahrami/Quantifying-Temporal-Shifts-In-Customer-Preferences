```python
total_population = table_2019[['A_cbg', 'E_Population_in_this_cbg']].drop_duplicates()[
    'E_Population_in_this_cbg'].sum()
total_population
```

```python
#total_population = tab
def get_weighted_distance(store_here):
    table_temp = table_2019[table_2019['B_store'] == store_here]
    table_temp['sum_distance'] = table_temp.apply(lambda x: x['E_Population_in_this_cbg'] * x[
        'G_Distance_between_cbg_and_store'], axis = 1)
    return table_temp['sum_distance'].sum()/total_population
```

```python
def get_visits(selected_stores_here, info_data):
    visits_2019_temp = table_2019[table_2019['B_store'].isin(selected_stores_here)].groupby(
        ['B_store'])['D_Number_of_Visits_2019'].sum().reset_index()
    visits_2020_temp = table_2020[table_2020['B_store'].isin(selected_stores_here)].groupby(
        ['B_store'])['D_Number_of_Visits_2020'].sum().reset_index()
    visits_2021_temp = table_2021[table_2021['B_store'].isin(selected_stores_here)].groupby(
        ['B_store'])['D_Number_of_Visits_2021'].sum().reset_index()
    
    final_df = visits_2019_temp.merge(visits_2020_temp).merge(visits_2021_temp).merge(
        info_data, how = 'left')[['O_Brand_name', 'B_store', 'J_POI_count_where_store_is', 
                                  'K_POI_diversity_where_store_is', 
                                  'H_Area_of_store', 'D_Number_of_Visits_2019', 'D_Number_of_Visits_2020', 
                                  'D_Number_of_Visits_2021']]
    final_df = final_df.rename(columns = {'O_Brand_name': 'Store Brand', 'B_store': 'Store ID', 
                                          'J_POI_count_where_store_is': 'CBG POI count', 
                                          'K_POI_diversity_where_store_is': 'CBG POI Diversity', 
                                          'H_Area_of_store': 'Area', 'D_Number_of_Visits_2019': 'Visits 2019', 
                                          'D_Number_of_Visits_2020': 'Visits 2020', 
                                          'D_Number_of_Visits_2021': 'Visits 2021'}).sort_values(
        by = 'Area', ascending = True)
    final_df['Weighted Distance'] = final_df['Store ID'].apply(lambda x: get_weighted_distance(x))
    this_store_brand = final_df[final_df['Store ID'] == selected_stores_here[0]]['Store Brand'].tolist()[0]
    for i in [2019, 2020, 2021]:
        final_df['share_' + str(i)] = final_df['Visits '+str(i)].apply(lambda x: x/final_df[
            final_df['Store Brand'] == this_store_brand]['Visits '+str(i)].sum())
        final_df['Visits '+str(i)] = final_df.apply(lambda x: str(x['Visits '+str(i)])+ ' (' + str(
            round(x['share_' + str(i)], 3)) + ')', axis = 1)
    return final_df[['Store Brand', 'Store ID', 'CBG POI count', 'CBG POI Diversity', 'Area', 
                     'Weighted Distance', 'Visits 2019', 'Visits 2020', 'Visits 2021']]
dollar_tree_1 = get_visits(['sg:ec8961ae058d42f684cc6bfe45e04aee', 'sg:7b03f24744384489bd8219d542d2ee7f'], 
                           dollar_tree_info)
dollar_tree_1
```
